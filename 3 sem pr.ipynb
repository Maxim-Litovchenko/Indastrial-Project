* Ð¡Ð´ÐµÐ»Ð°Ñ‚ÑŒ RL Q-Learning Ð¼Ð¾Ð´ÐµÐ»ÑŒ, Ñ Ð½Ð°Ð³Ñ€Ð°Ð´Ð¾Ð¹ Ð·Ð° Ð¿Ð¾Ð´Ð´ÐµÑ€Ð¶ÐºÑƒ Ñ†ÐµÐ»ÐµÐ²Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð² Ð·Ð°Ð´Ð°Ð½Ð½Ð¾Ð¼ Ð´Ð¸Ð°Ð¿Ð¾Ð·Ð»Ð½Ðµ Ð¸ Ð²Ð¾Ð·Ð¼Ð¾Ð¶Ð½Ð¾ÑÑ‚ÑŒÑŽ Ð¼ÐµÐ½ÑÑ‚ÑŒ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ ÑƒÑÑ‚Ð°Ð²Ð¾Ðº
Q-Learning Ð±Ñ‹Ð» Ð²Ñ‹Ð±Ñ€Ð°Ð½, ÐºÐ°Ðº ÑÐ°Ð¼Ð°Ñ Ð¿Ñ€Ð¾ÑÑ‚Ð°Ñ Ð¼Ð¾Ð´ÐµÐ», Ð¸Ð·-Ð·Ð° Ð¾Ð³Ñ€Ð°Ð½Ð¸Ñ‡ÐµÐ½Ð¸Ñ Ð¿Ð¾ Ñ€ÐµÑÑƒÑ€ÑÐ°Ð¼   

```python
import random
from sklearn.metrics import mean_absolute_error

class QLearningAgent:
    def __init__(self, state_space_size, action_space_size, alpha=0.1, gamma=0.99, epsilon=0.1):
        self.state_space_size = state_space_size
        self.action_space_size = action_space_size
        self.alpha = alpha  
        self.gamma = gamma 
        self.epsilon = epsilon 
        self.q_table = {}
        self.discrete_values = [-0.05, 0, 0.05]  

    def get_q_value(self, state, action):
        return self.q_table.get((state, action), 0.0)
    
    def choose_action(self, state):
        if random.uniform(0, 1) < self.epsilon:
            return random.choice(range(self.action_space_size))
        else:
            q_values = [self.get_q_value(state, action) for action in range(self.action_space_size)]
            return np.argmax(q_values)
    
    def update_q_table(self, state, action, reward, next_state):
        best_next_action = self.choose_action(next_state)
        td_target = reward + self.gamma * self.get_q_value(next_state, best_next_action)
        td_error = td_target - self.get_q_value(state, action)
        self.q_table[(state, action)] = self.get_q_value(state, action) + self.alpha * td_error

def discretize_state(state, bins):
    return tuple(np.digitize(s, b) - 1 for s, b in zip(state, bins))

n_bins = 10  
state_columns = ['Ni1', 'S1', 'Fe3O41', 'Ni2', 'S2', 'Fe3O42',
       'Ni3', 'S3', 'Fe3O43', 'Ni4', 'S4', 'Fe3O44', 'Ni_final', 'S_final',
       'Fe3O4_final', 'ro1', 'ro2', 'ro3', 'ro4', 'ro_final', 'ust1', 'ust2',
       'ust3', 'ust4', 'v_final', 'level_final', 'time_diff', 'is_duplicate']
state_bins = [np.linspace(data[col].min(), data[col].max(), n_bins) for col in state_columns]
state_space_size = (n_bins,) * len(state_columns)
action_space_size = len(agent.discrete_values) ** 4  

agent = QLearningAgent(state_space_size=state_space_size, action_space_size=action_space_size)

n_episodes = 20  
rewards = []

for episode in range(n_episodes):
    state_index = random.randint(0, len(data) - 2)  
    state = discretize_state(data.iloc[state_index][state_columns].values, state_bins)
    total_reward = 0
    
    for t in range(state_index, len(data) - 1):
        action = agent.choose_action(state)
        
        idx1, idx2, idx3, idx4 = np.unravel_index(action, (3, 3, 3, 3))
        change_ust1 = agent.discrete_values[idx1]
        change_ust2 = agent.discrete_values[idx2]
        change_ust3 = agent.discrete_values[idx3]
        change_ust4 = agent.discrete_values[idx4]
        
        next_state_raw = data.iloc[t + 1].copy()
        next_state_raw['ust1'] += change_ust1
        next_state_raw['ust2'] += change_ust2
        next_state_raw['ust3'] += change_ust3
        next_state_raw['ust4'] += change_ust4
        
        next_state = discretize_state(next_state_raw[state_columns].values, state_bins)
        
        ro = next_state_raw['ro_final']
        Ni = next_state_raw['Ni_final']
        S = next_state_raw['S_final']
        Fe3O4 = next_state_raw['Fe3O4_final']

        ro_reward = 1 if 1.85 <= ro <= 2.5 else -abs(ro - 2.175)
        Ni_reward = 1 if 7.275 <= Ni <= 7.725 else -abs(Ni - 7.5)
        S_reward = 1 if 24.735 <= S <= 26.265 else -abs(S - 25.5)
        Fe3O4_reward = 1 if 1.649 <= Fe3O4 <= 1.751 else -abs(Fe3O4 - 1.7)

        reward = ro_reward + Ni_reward + S_reward + Fe3O4_reward
        
        agent.update_q_table(state, action, reward, next_state)
        state = next_state
        total_reward += reward
        
        if t == len(data) - 2:
            done = True
        else:
            done = False
        
        if done:
            break
    
    rewards.append(total_reward)  
    
    if episode % 1 == 0:
        print(f'Episode {episode}: Total Reward: {total_reward}')

import pickle
with open('q_table.pkl', 'wb') as f:
    pickle.dump(agent.q_table, f)

mean_absolute_error_value = mean_absolute_error(rewards, [4] * len(rewards))
print(f'Mean Absolute Error: {mean_absolute_error_value:.4f}')
```
